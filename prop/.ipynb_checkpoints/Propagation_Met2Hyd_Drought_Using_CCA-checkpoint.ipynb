# Propagation Meteorological to Hydrological Drought using Cross-Correlation

The notebook explain on how to analyzing the propagation of **meteorological drought** (Standardized Precipitation Index - SPI) to **hydrological drought** (Standardized Streamflow Index - SSI) using **cross-correlation analysis** at the pixel level. It required two input netCDF files, SPI - generated from `Steps_to_Generate_SPI_Using_CHIRPS_Data.ipynb` and SSI - generated from `Steps_to_Generate_SSI_Using_GloFAS-ERA5_Data.ipynb`.

As background, we have SPI and SSI data which has same temporal resolution (monthly) and cover the same period (1981-2022), also has same various time scale: `1`, `2`, `3`, `6`, `9`, `12`, `18`, `24`, `36`, `48`, `60` and `72-month`. 


The analysis **ideally** will take some process.

1. **Choice of Time Scale:**

    Let's start with the 6-month timescale. This timescale is long enough to smooth out short-term variability but short enough to capture seasonal patterns.

2. **Handling Seasonality:**

    While splitting the dataset by season can be complex due to varying wet and dry seasons across regions, we might consider a simpler approach: analyze the entire dataset first, then conduct a separate analysis for regions with different climatic patterns. Another method is to detrend the data to remove seasonality, but this might also remove some meaningful variability related to the drought dynamics. **For simplicity, we will not do this part.**

3. **Noise Filtering:**

    Employing noise filtering techniques like Singular Spectrum Analysis (SSA) or a moving average can indeed help in isolating the underlying trends and patterns in our data. This step is crucial for enhancing the signal-to-noise ratio in our datasets.

4. **Cross-Correlation Analysis:**

    Perform the cross-correlation for each pixel across the entire time series. This will involve computing the correlation for different lags (e.g., 1 month, 2 months, up to a reasonable maximum that makes sense for our study).

5. **Output and Visualization**

    **Lag Map:** This will be particularly insightful, showing the spatial distribution of the lag between meteorological and hydrological droughts.

    **Strength Map:** This map will be useful in understanding where the correlations are strongest, indicating regions where meteorological conditions are good predictors of hydrological conditions.


## 1. Data Loading and Inspection

This step involves loading Standardized Precipitation Index (SPI) and Standardized Streamflow Index (SSI) datasets, crucial for assessing meteorological and hydrological droughts. Key aspects of the script include:

* **Data Access:** Retrieves SPI and SSI datasets from specified file paths.
* **Dimension Extraction:** Identifies the number of latitude and longitude points, offering spatial context.
* **Time Selection:** Focuses on data from June 1981 to December 2022, ensuring temporal relevance.
* **Data Transformation:** Converts xarray Datasets to pandas DataFrames for comprehensive analysis.
* **NaN Handling:** Implements a function to summarize and report on NaN values in the datasets, maintaining data integrity.
* **Land Area Masking:** Masks non-land areas, identified by consistent NaN values, enhancing the focus on relevant land regions.
* **Data Summary:** Provides an updated overview post-masking, offering insights into the remaining NaN values and data completeness.

This preparatory phase establishes a solid foundation for subsequent analytical steps, ensuring data accuracy and relevance.

```python
import xarray as xr
import numpy as np
import pandas as pd

print("Loading SPI and SSI data...")
# Load the SPI and SSI data
spi_dataset = xr.open_dataset('../met_drought/05_spi/idn_cli_spi_gamma_6_month.nc')
ssi_dataset = xr.open_dataset('../hyd_drought/08_ssi/idn_cli_ssi_gamma_6_month.nc')

# Extract latitude and longitude dimensions
print("Extracting latitude and longitude dimensions...")
num_lat = len(spi_dataset['lat'])  # Replace 'lat' with the actual latitude dimension name
num_lon = len(spi_dataset['lon'])  # Replace 'lon' with the actual longitude dimension name

print("Number of Latitude Points:", num_lat)
print("Number of Longitude Points:", num_lon)

print("Selecting data from June 1981 to December 2022...")
# Selecting data from June 1981 to December 2022
spi_selected = spi_dataset.sel(time=slice('1981-06-01', '2022-12-31'))
ssi_selected = ssi_dataset.sel(time=slice('1981-06-01', '2022-12-31'))

print("Converting xarray Datasets to pandas DataFrames...")
# Convert the xarray Datasets to pandas DataFrames
spi_df = spi_selected.to_dataframe().unstack('time')['spi_gamma_6_month']
ssi_df = ssi_selected.to_dataframe().unstack('time')['ssi_gamma_6_month']

# Function to summarize NaNs in a DataFrame
def summarize_nans(df):
    nan_counts = df.isna().sum()
    total_values = df.size
    total_nans = nan_counts.sum()
    print(f"Total NaNs: {total_nans}")
    if total_nans > 0:
        print(f"Percentage of NaNs: {total_nans / total_values * 100:.2f}%")
    else:
        print("No NaN values present.")
    return nan_counts

# Revisit the masking logic
def mask_non_land_areas(df):
    """
    Mask the non-land areas in the DataFrame.
    Non-land areas are identified as spatial points that have NaNs for all time points.
    """
    print("Masking non-land areas...")
    # Find spatial points with at least one non-NaN value across all times
    valid_spatial_points = df.notna().any(axis=1)

    # Filter the DataFrame to keep only valid spatial points
    return df[valid_spatial_points]

print("Applying masking to original DataFrames...")
# Apply the masking to the original DataFrames
spi_df_land_only = mask_non_land_areas(spi_df)
ssi_df_land_only = mask_non_land_areas(ssi_df)

print("Rechecking NaN summary for the masked data...")
# Recheck NaN summary for the masked data
print("SPI Data Summary (Land Only):")
spi_nan_summary_land_only = summarize_nans(spi_df_land_only)
print("\nSSI Data Summary (Land Only):")
ssi_nan_summary_land_only = summarize_nans(ssi_df_land_only)
```

## 2. Noise Filtering with Singular Spectrum Analysis

This part of our analysis focuses on applying Singular Spectrum Analysis (SSA) for noise filtering and trend extraction in Standardized Precipitation Index (SPI) and Standardized Streamflow Index (SSI) datasets. SSA is particularly adept at separating the signal (trend) from the noise in time series data. Key steps and features of this script include:

* **SSA Implementation:** Utilizes the `seasonal_decompose` function from the `statsmodels` library for SSA, effectively decomposing time series data into trend, seasonal, and residual components.
* **Data Preparation and Processing:**
    * Converts xarray DataArrays to pandas DataFrames for SSA application.
    * Handles missing values in SSI data through forward filling, ensuring data continuity.
* **Trend Extraction:** Focuses on extracting the trend component from both SPI and SSI data, considered as the crucial signal of interest.
* **Data Transformation:** Transforms the extracted trend components back into xarray DataArrays for further analysis.
* **Metadata Assignment:** Assigns appropriate metadata to the trend components, enhancing data understanding and context.
* **NetCDF File Output:** Saves the filtered datasets as NetCDF files, adhering to the Climate and Forecast (CF) Conventions, making them ready for subsequent analysis or visualization.

```python
import pandas as pd
from statsmodels.tsa.seasonal import seasonal_decompose
from tqdm.auto import tqdm
import xarray as xr
import os

# Enable pandas progress_apply with tqdm
tqdm.pandas()

def apply_ssa(data, freq):
    """
    Apply Singular Spectrum Analysis (SSA) to extract the trend component.
    """
    result = seasonal_decompose(data, model='additive', period=freq)
    return result.trend

print("Starting SSA process...")

# Apply SSA to SPI data
print("Applying SSA to SPI data...")
spi_trend = spi_df_land_only.progress_apply(lambda x: apply_ssa(x, freq=12), axis=0)
print("SSA applied to SPI data.")

# Handle NaNs in SSI data before applying SSA
print("Handling NaNs in SSI data...")
ssi_df_land_only_filled = ssi_df_land_only.fillna(method='ffill')  # Forward fill NaNs
print("NaNs handled. Applying SSA to SSI data...")
ssi_trend = ssi_df_land_only_filled.progress_apply(lambda x: apply_ssa(x, freq=12), axis=0)
print("SSA applied to SSI data.")

print("Converting filtered data to xarray DataArrays...")

# Convert the multi-indexed DataFrame to xarray Dataset
spi_trend_ds = spi_trend_multi.to_xarray()
ssi_trend_ds = ssi_trend_multi.to_xarray()

# Combine all variables into a single DataArray
spi_trend_da = spi_trend_ds.to_array(dim='time')
ssi_trend_da = ssi_trend_ds.to_array(dim='time')

# Align the time dimension (skip the renaming step)
spi_trend_da = spi_trend_da.reindex_like(spi_dataset, method='nearest')
ssi_trend_da = ssi_trend_da.reindex_like(ssi_dataset, method='nearest')

print("Assigning coordinates and metadata...")
spi_trend_da = spi_trend_da.assign_coords(time=spi_dataset['time'].values)
ssi_trend_da = ssi_trend_da.assign_coords(time=ssi_dataset['time'].values)

# Check if the dimensions match
print("SPI DataArray dimensions:", spi_trend_da.dims)
print("SSI DataArray dimensions:", ssi_trend_da.dims)

print("Creating xarray Datasets and adding attributes...")
# Create xarray Datasets and add attributes
spi_trend_ds = xr.Dataset({'spi_filtered': spi_trend_da})
spi_trend_ds.spi_filtered.attrs = {
    'standard_name': 'standardized_precipitation_index',
    'long_name': 'Standardized Precipitation Index (Gamma), 6-month (Filtered)',
    'units': '1',  # SPI is unitless
    'comment': 'Filtered using Singular Spectrum Analysis'
}
ssi_trend_ds = xr.Dataset({'ssi_filtered': ssi_trend_da})
ssi_trend_ds.ssi_filtered.attrs = {
    'standard_name': 'standardized_streamflow_index',
    'long_name': 'Standardized Streamflow Index (Gamma), 6-month (Filtered)',
    'units': '1',  # SSI is unitless
    'comment': 'Filtered using Singular Spectrum Analysis'
}

# Save the Datasets as NetCDF files following CF Conventions
def save_dataset(dataset, filename):
    if os.path.exists(filename):
        print(f"File '{filename}' already exists and will be replaced.")
    dataset.to_netcdf(filename)

print("Saving the datasets as NetCDF files...")
save_dataset(spi_trend_ds, 'idn_cli_spi_gamma_6_month_filtered.nc')
save_dataset(ssi_trend_ds, 'idn_cli_ssi_gamma_6_month_filtered.nc')
print("Datasets saved as NetCDF files.")

print("Process completed. Displaying dataset summaries:")
# Check the result (optional)
print(spi_trend_ds)
print(ssi_trend_ds)
```

## 3: Cross-Correlation Analysis

This script performs a detailed cross-correlation analysis between meteorological (SPI) and hydrological (SSI) drought indices over a geographical region. Key aspects include:

* **Loading Data:* Imports meteorological and hydrological drought indices from NetCDF files.
* **Land Mask Application:** Utilizes a land mask to focus analysis on terrestrial areas, excluding ocean or sea regions.
* **Cross-Correlation Calculation:**
    * Computes the correlation for various time lags (up to 12 months).
    * Analyzes each grid cell separately, ensuring detailed spatial analysis.
* **Result Processing:** Identifies the time lag with the maximum correlation at each location, providing insights into drought propagation.
* **Data Array Creation:** Constructs xarray DataArrays for both lag time and correlation strength, including appropriate metadata.
* **NetCDF Output:** Saves the processed data as a NetCDF file, suitable for further analysis or visualization.

```python
import xarray as xr
import numpy as np
from tqdm import tqdm
import os

# Load the land mask data from NetCDF
land_mask_ds = xr.open_dataset('idn_subset_chirps.nc')
land_mask = land_mask_ds['land'].astype(bool)  # Assuming land points are marked with 1

def calculate_cross_correlation(spi_data, ssi_data, land_mask, max_lag):
    print("Starting cross-correlation calculation...")
    lags = np.arange(1, max_lag + 1)
    lat_len, lon_len = spi_data.shape[1], spi_data.shape[2]
    corr_map = np.full((lat_len, lon_len, max_lag), np.nan)

    pbar = tqdm(total=lat_len * lon_len, desc='Calculating Cross-Correlation')
    for lat_idx in range(lat_len):
        for lon_idx in range(lon_len):
            # Debugging output
            if lat_idx % 100 == 0 and lon_idx == 0:
                print(f"Processing lat index: {lat_idx}")
            
            # Proceed only if the current grid cell is land
            if land_mask[lat_idx, lon_idx]:
                spi_series = spi_data[:, lat_idx, lon_idx]
                ssi_series = ssi_data[:, lat_idx, lon_idx]
                valid_spi = ~np.isnan(spi_series)
                valid_ssi = ~np.isnan(ssi_series)
                valid_idx = valid_spi & valid_ssi

                if np.sum(valid_idx) > 0:
                    for lag in lags:
                        if np.sum(valid_idx) > lag:
                            corr = np.corrcoef(spi_series[valid_idx][:-lag], ssi_series[valid_idx][lag:])[0, 1]
                            corr_map[lat_idx, lon_idx, lag - 1] = corr
            else:
                # Explicitly set sea points to NaN for all lags
                corr_map[lat_idx, lon_idx, :] = np.nan
            pbar.update(1)
    pbar.close()
    return corr_map

print("Loading filtered data from NetCDF files...")
spi_ds = xr.open_dataset('idn_cli_spi_gamma_6_month_filtered.nc')
ssi_ds = xr.open_dataset('idn_cli_ssi_gamma_6_month_filtered.nc')

print("Extracting DataArrays...")
spi_data = spi_ds['spi_filtered']
ssi_data = ssi_ds['ssi_filtered']

max_lag = 12

# Assuming spi_data and ssi_data are your DataArray objects from the NetCDF files
land_mask_aligned = land_mask.reindex_like(spi_data, method='nearest')

# Now pass this aligned land mask to your cross-correlation function
corr_map = calculate_cross_correlation(spi_data.values, ssi_data.values, land_mask_aligned.values, max_lag)

print("Processing results...")
# Initialize an array full of NaNs for the lag
max_corr_lag = np.full(corr_map.shape[:2], np.nan)

# Loop over each point in the correlation map
for lat_idx in range(max_corr_lag.shape[0]):
    for lon_idx in range(max_corr_lag.shape[1]):
        # Only consider points that are on land
        if land_mask[lat_idx, lon_idx]:
            # Extract the slice of the correlation map for this point
            corr_slice = corr_map[lat_idx, lon_idx, :]

            # If not all values are NaN, find the index of the max correlation
            if not np.all(np.isnan(corr_slice)):
                max_idx = np.nanargmax(corr_slice)
                max_corr_lag[lat_idx, lon_idx] = max_idx + 1  # Add 1 because index starts at 0

# The max_corr_value can be computed as before since it's independent of the lag index
max_corr_value = np.full_like(max_corr_lag, np.nan, dtype=np.float32)
for lat_idx in range(max_corr_lag.shape[0]):
    for lon_idx in range(max_corr_lag.shape[1]):
        if not np.all(np.isnan(corr_map[lat_idx, lon_idx, :])):
            max_corr_value[lat_idx, lon_idx] = np.nanmax(corr_map[lat_idx, lon_idx, :])

# Create DataArrays with appropriate attributes
max_corr_lag_da = xr.DataArray(
    max_corr_lag,
    coords={'lat': spi_data.lat, 'lon': spi_data.lon},
    dims=["lat", "lon"],
    name='max_corr_lag',
    attrs={
        'long_name': 'Lag Time with Maximum Correlation',
        'units': 'months',
        'description': 'Lag time where the maximum correlation between SPI and SSI is observed'
    }
)

max_corr_value_da = xr.DataArray(
    max_corr_value,
    coords={'lat': spi_data.lat, 'lon': spi_data.lon},
    dims=["lat", "lon"],
    name='max_corr_value',
    attrs={
        'long_name': 'Maximum Correlation Value',
        'units': '',
        'description': 'Maximum correlation value between SPI and SSI'
    }
)

# Set attributes for latitude and longitude coordinates
max_corr_lag_da.lat.attrs = {'standard_name': 'latitude', 'units': 'degrees_north'}
max_corr_lag_da.lon.attrs = {'standard_name': 'longitude', 'units': 'degrees_east'}
max_corr_value_da.lat.attrs = {'standard_name': 'latitude', 'units': 'degrees_north'}
max_corr_value_da.lon.attrs = {'standard_name': 'longitude', 'units': 'degrees_east'}

# Combine DataArrays into a Dataset
ds = xr.Dataset({
    'max_corr_lag': max_corr_lag_da,
    'max_corr_value': max_corr_value_da
})

# Add global attributes to the Dataset
ds.attrs = {
    'title': 'Drought Propagation from Meteo to Hydro using Cross-Correlation Analysis, 6-month time scale',
    'institution': 'Department of Geophysics and Meteorology, IPB - Bogor, Indonesia',
    'source': 'Climate Analysis',
    'author': 'Benny Istanto',
    'history': f'Created on {pd.Timestamp.now()}',
    'Conventions': 'CF-1.8'
}

# Save to netcdf
def save_to_netcdf(ds, filename):
    try:
        if os.path.exists(filename):
            os.remove(filename)  # Attempt to remove the file if it already exists
            print(f"Existing file '{filename}' has been removed.")
        encoding = {'max_corr_lag': {'dtype': 'int16'}, 'max_corr_value': {'dtype': 'float32'}}
        ds.to_netcdf(filename, mode='w')
        print(f"Dataset has been saved to '{filename}'.")
    except PermissionError as e:
        print(f"PermissionError: {e}")
    except Exception as e:
        print(f"An error occurred: {e}")

filename = 'idn_cli_droughtprop_06_scale_analysis.nc'
print(f"Attempting to save to NetCDF file: {filename}")
save_to_netcdf(ds, filename)

print('Process completed!')
```

## 4. Visualisation

The visualization script is designed to illustrate the results of the cross-correlation analysis between meteorological (SPI) and hydrological (SSI) droughts. Key features of the script include:

* **Lag Map Creation:** This map displays the time lag (in months) between meteorological and hydrological droughts across the study area. It helps identify regions where hydrological responses to meteorological changes are immediate or delayed.

* **Strength Map Generation:** This map shows the strength of the correlation between meteorological and hydrological droughts. It highlights areas with a strong predictive relationship, indicating regions sensitive to meteorological changes.

* **Spatial Insight:** Both maps together provide a spatial understanding of drought dynamics, highlighting areas of resilience and vulnerability.

* **Practical Application:** The visualizations aid in drought management by pinpointing critical areas for focused attention and intervention.


```python
import matplotlib.pyplot as plt
import matplotlib.colors as colors
import cartopy.crs as ccrs
import numpy as np
import xarray as xr

def plot_map(data_array, title, cmap, n_classes, cbar_label='', save_filename=None, is_lag_map=False, title_font_size=10):
    plt.figure(figsize=(10, 6))
    ax = plt.axes(projection=ccrs.PlateCarree())
    ax.coastlines()

    # Set extent based on the data coordinates
    extent = [data_array.lon.min(), data_array.lon.max(), data_array.lat.min(), data_array.lat.max()]
    ax.set_extent(extent, crs=ccrs.PlateCarree())

    # Define the colormap and norm to create discrete intervals
    cmap = plt.get_cmap(cmap, n_classes)

    # Set boundaries for discrete intervals
    if is_lag_map:
        boundaries = np.arange(1, n_classes + 2)  # For Lag Map: 1 to 12
    else:
        boundaries = np.linspace(0.1, 1.0, n_classes + 1)  # For Strength Map: 0.1 to 1.0

    norm = colors.BoundaryNorm(boundaries=boundaries, ncolors=n_classes)

    # Plot the data
    im = ax.pcolormesh(data_array.lon, data_array.lat, data_array, cmap=cmap, norm=norm, transform=ccrs.PlateCarree())

    # Set title with specified font size
    ax.set_title(title, fontsize=title_font_size)
    
    # Place the colorbar at the bottom of the map and set the ticks
    cbar = plt.colorbar(im, orientation='horizontal', fraction=0.05, pad=0.05, aspect=30)
    cbar.set_label(cbar_label)

    # Set tick labels for discrete intervals
    if is_lag_map:
        tick_positions = np.arange(1.5, n_classes + 1.5)
        tick_labels = np.arange(1, n_classes + 1)
    else:
        tick_positions = np.linspace(0.15, 0.95, n_classes)
        tick_labels = ['{:.1f}'.format(val) for val in np.linspace(0.1, 1.0, n_classes)]

    cbar.set_ticks(tick_positions)
    cbar.set_ticklabels(tick_labels)

    if save_filename:
        plt.savefig(save_filename, dpi=300, bbox_inches='tight')
    plt.show()

# Load data from NetCDF file
nc_data = xr.open_dataset('idn_cli_droughtprop_06_scale_analysis.nc')
max_corr_lag_da = nc_data['max_corr_lag']
max_corr_value_da = nc_data['max_corr_value']

# Plotting the maps with discrete classes and customized colorbar ticks
plot_map(max_corr_lag_da, 'The Time lag (in months) between meteorological and hydrological droughts, using 6-month timescale data', 
         cmap='coolwarm', n_classes=12, cbar_label='Lag (months)', title_font_size=10,
         save_filename='./01_cca/images/idn_cli_droughtprop_cca_06_scale_lag_map.png', is_lag_map=True)
plot_map(max_corr_value_da, 'The Strength of the correlation between meteorological and hydrological droughts, using 6-month timescale data', 
         cmap='inferno', n_classes=10, cbar_label='Correlation Strength', title_font_size=10,
         save_filename='./01_cca/images/idn_cli_droughtprop_cca_06_scale_strength_map.png')
```
